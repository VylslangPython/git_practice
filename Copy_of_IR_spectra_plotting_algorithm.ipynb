```python
#@title Importing modules { form-width: "10%" }
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#@title File { form-width: "10%" }


filename = "Excel_CB_06_3"

compound_name = "Dibenzyldimethylammonium bromide"

# Enter desired frequencies for specific plot
#NOTE: higher frequencies FIRST
starting_frequency_1 = 1600
ending_frequency_1 = 700

starting_frequency_2 = 3300
ending_frequency_2 = 2600


# zoomed_out => first two plots
linewidth_zoomed_out = 0.55
linewidth_zoomed_in = 1.1
```

```python
#@title Reading main file { form-width: "20%" }


spectra_og = pd.read_excel(f"{filename}.xlsx", skiprows = 1)

# print([[spectra_og]]) #double square brackets => prints as dataset as opposed to pandas series object
```

```python
#@title Data tests { form-width: "20%" }


def data_reading_test():
  print("Element:", spectra_og["cm-1"][4], "\nDatatype:", type(spectra_og["cm-1"][4]))

def dtype_check():
  ls = []
  for i in spectra_og["cm-1"]:
    if isinstance(i, float) == False: #checks if they're all floats
      ls.append(i)

  for i in spectra_og["%T"]:
    if isinstance(i, float) == False: #checks if they're all floats
      if not i.isdecimal():
        ls.append(type(i))
        
  print("ls", ls)

# data_reading_test()
# dtype_check()
```

```python
#@title Plotting data { form-width: "20%" }

frequency = list(spectra_og["cm-1"])
transmittance = list(spectra_og["%T"])

frequency_unit = "$\mathrm{cm}^{-1}$"


plt.figure(figsize=(8, 6), dpi=100)
plt.plot(frequency, transmittance, linewidth=linewidth_zoomed_out)
plt.xlim(max(frequency), min(frequency))
plt.title(f"IR spectra of {compound_name} (All measured data)")
plt.xlabel(f"Frequency ({frequency_unit})")
plt.ylabel("Transmittance (%)")
plt.show()


#making a new pectrum with specific frequencies
frequency = list(spectra_og["cm-1"][:-200])
transmittance = list(spectra_og["%T"][:-200])

plt.figure(figsize=(8, 6), dpi=100)
plt.title(f"\n\nIR spectra of {compound_name} (Relevant frequencies)")
plt.xlabel(f"Frequency ({frequency_unit})")
plt.ylabel("Transmittance (%)")
plt.plot(frequency, transmittance, linewidth=linewidth_zoomed_out)
plt.xlim(max(frequency), min(frequency))
plt.show()



# comparison to reference spectrum for specified wavelengths
#plot 1
frequency = list(spectra_og["cm-1"][4000 - starting_frequency_1 : -ending_frequency_1 + 450])
transmittance = list(spectra_og["%T"][4000 - starting_frequency_1 : -ending_frequency_1 + 450])

plt.figure(figsize=(8, 6), dpi=100)
plt.plot(frequency, transmittance, linewidth=linewidth_zoomed_in)
plt.xlim(max(frequency), min(frequency))
plt.title(f"\n\nIR spectra of {compound_name} \n(Specific region of {starting_frequency_1} to {ending_frequency_1} {frequency_unit} for comparison with reference spectrum)")
plt.xlabel(f"Frequency ({frequency_unit})")
plt.ylabel("Transmittance (%)")
plt.show()

#plot 2
frequency = list(spectra_og["cm-1"][4000 - starting_frequency_2 : -ending_frequency_2 + 450])
transmittance = list(spectra_og["%T"][4000 - starting_frequency_2 : -ending_frequency_2 + 450])

plt.figure(figsize=(8, 6), dpi=100)
plt.plot(frequency, transmittance, linewidth=linewidth_zoomed_in)
plt.xlim(max(frequency), min(frequency))
plt.title(f"\n\nIR spectra of {compound_name} \n(Specific region of {starting_frequency_2} to {ending_frequency_2} {frequency_unit} for comparison with reference spectrum)")
plt.xlabel(f"Frequency ({frequency_unit})")
plt.ylabel("Transmittance (%)")
plt.show()
```
